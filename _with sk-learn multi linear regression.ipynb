# step 2: training the LinearRegression Model




```python
from sklearn.model_selection import train_test_split
```

```python
#  4 input variables for multiple regression
X = data[['Pregnancies', 'Glucose', 'BMI', 'DiabetesPedigreeFunction']]
y = data['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=101)
```

```python
#import linearregression
from sklearn.linear_model import LinearRegression

reg= LinearRegression()
reg.fit(X_train,y_train)
```

```python
print("Intercept value = ",reg.intercept_)
print('Coefficients value = ')
coefficient = pd.DataFrame(reg.coef_,X.columns,columns=['Coefficient'])
coefficient
```

```python
from matplotlib import pyplot as plt
coefficient['Coefficient'].plot(kind='line', figsize=(8, 4), title='Coefficient')
plt.gca().spines[['top', 'right']].set_visible(False)
```

 <table><tr><font color= blodwhite   size=4 face="微软雅黑">
 step 3: Predictions from our Linear Regression Model

```python
y_pred = reg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test,y_pred) #

print('prediction: ',score)
print('Coefficients: ', reg.coef_)
print("Mean squared error: %.2f" % np.mean((reg.predict(X_test) - y_test) ** 2))
print('Variance score: %.2f' % reg.score(X_test, y_test))
```

```python
plt.scatter(y_test,y_pred, color='green')
plt.title('the result', fontsize=14)
plt.xlabel('actual result', fontsize=11)
plt.ylabel('predict result', fontsize=11)
plt.show()
```
